```python
from torch.utils.data import Dataset
import cv2
import numpy as np
class HUBMAPDataset(Dataset):
    def __init__(self,ids,transform=None):
        self.ids = ids
        self.transform = transform
        
    def __len__(self):
        return len(self.ids)
    def __getitem__(self,idx):
        name = self.ids[idx]
        image_name = name.split('_')[0]
        img = cv2.imread(f'../../data/processed/train/{image_name}/{name}.png').astype('float32')
        img /= 255.
        mask = cv2.imread(f'../../data/processed/train/{image_name}/{name}_mask.png',0)
        label = (mask.sum()>0)*1
        if self.transform is None:
            img = img.transpose(2,0,1).astype('float32')
            mask = mask.transpose(2,0,1).astype('float32')
            return {'image': img,
                    'mask': mask,
                   'label':label,
                   'name':name}
        else:

            img = self.transform()(image=img,mask=mask)["image"]
            mask = self.transform()(image=img,mask=mask[:,:,np.newaxis])["mask"]
            
            img = img.transpose(2,0,1).astype('float32')
            mask = mask.transpose(2,0,1).astype('float32')
            
            return {'image': img,
                  'mask': mask,
                  'label':label,
                   'name':name}
```

```python
import pickle
from torch.utils.data import DataLoader
train_files_name = pickle.load(open('../result/seresnext101/trn_idxs_list_seed1024.pickle','rb'))
# train and val data transforms:
from transform import get_train_transforms
for fold in [0,1,2,3]:
    print(f'training on fold{fold}')

# datasets:
    train_set = HUBMAPDataset(train_files_name[fold],get_train_transforms)
#     val_set = dataset.HUBMAPDataset(val_files_name[fold],get_valid_transforms)

    # dataloaders:
    train_loader = DataLoader(train_set, batch_size=8, shuffle=True, num_workers=1)
#     val_loader = DataLoader(val_set, batch_size=val_batch_size, shuffle=False, num_workers=num_workers)
    break
```

```python
def plot_epoch_samples(batchsize,images,masks,predictions,save_name):
    image2=images.numpy().transpose(0,2,3,1)
    image = np.zeros(image2[0].shape)
    mask2 = masks.numpy().transpose(0,2,3,1)
    mask = np.zeros(mask2[0].shape)
    prediction2 = predictions.numpy().transpose(0,2,3,1)
    prediction = np.zeros(prediction2[0].shape)
    for i in range(batchsize):
        image = np.concatenate([image,image2[i]],axis=1)
        mask = np.concatenate([mask,mask2[i]],axis=1)
        prediction = np.concatenate([prediction,prediction2[i]],axis=1)
    plt.imshow((mask[:,mask2.shape[2]:,0]*255).astype(np.uint8))
    plt.show()
    contours,_ = cv2.findContours((mask[:,mask2.shape[2]:,:]*255).astype(np.uint8),cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    contours_pred,_ = cv2.findContours((prediction[:,prediction2.shape[2]:,:]*255).astype(np.uint8),cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    img = cv2.cvtColor(np.array(image[:,image2.shape[2]:,:]*255,dtype=np.uint8), cv2.COLOR_RGB2BGR)
#     plt.imshow(img)
    img2= cv2.drawContours(img,contours,-1,(0,255,0),3)
    plt.imshow(img2)
    plt.show()
    img2= cv2.drawContours(img2,contours_pred,-1,(0,0,255),3)
    cv2.imwrite(save_name,img2)

plot_epoch_samples(8,image,mask,mask,'test.png')
```
